```python
# Memuat dan memproses data
def muat_dan_proses_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print(f"Berhasil membaca file: {file_path}")
        print(f"Jumlah baris awal: {len(df)}")
        
        # Preprocessing data
        df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
        df['Year'] = df['Date'].dt.year
        df['Month'] = df['Date'].dt.month
        df['Day'] = df['Date'].dt.day
        df['DayOfWeek'] = df['Date'].dt.dayofweek
        
        # Menambahkan fitur lag untuk penjualan sebelumnya
        df['Sales_Lag1'] = df.groupby('Store')['Weekly_Sales'].shift(1)
        df['Sales_Lag2'] = df.groupby('Store')['Weekly_Sales'].shift(2)
        
        # Menghitung rata-rata penjualan per toko
        df['Avg_Store_Sales'] = df.groupby('Store')['Weekly_Sales'].transform('mean')
        
        # Menghitung safety stock (contoh: 20% dari rata-rata penjualan)
        df['Safety_Stock'] = df['Avg_Store_Sales'] * 0.2
        
        # Menghitung target: Required Stock
        df['Required_Stock'] = df['Weekly_Sales'] + df['Safety_Stock']
        
        # Hapus baris dengan NaN yang muncul karena penggunaan lag
        df.dropna(inplace=True)
        print(f"Jumlah baris setelah menghapus NaN: {len(df)}")
        
        # Hapus kolom yang tidak diperlukan
        df.drop(['Date', 'Weekly_Sales'], axis=1, inplace=True)
        
        return df
    except Exception as e:
        print(f"Terjadi kesalahan saat memuat atau memproses data: {e}")
        return None

file_path = r"C:\Users\mp22r\walmart-sales\Walmart_sales.csv"
df = muat_dan_proses_data(file_path)

if df is not None:
    # Analisis Data Eksploratif (EDA)
    print("\nAnalisis Data Eksploratif (EDA)")
    print(df.describe())
    print(df.info())

    # Visualisasi data
    def visualize_data(df):
        plt.figure(figsize=(12, 6))
        sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
        plt.title('Correlation Heatmap')
        plt.show()

    visualize_data(df)

# Membagi dan menormalisasi data
def bagi_dan_normalisasi_data(df, target_col='Required_Stock', test_size=0.2, random_state=42):
    X = df.drop(target_col, axis=1)
    y = df[target_col]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, y_train, y_test, scaler

X_train, X_test, y_train, y_test, scaler = bagi_dan_normalisasi_data(df)
```
